```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_raw = sns.load_dataset("mpg")
df = df_raw.describe().T.loc[:, ['mean', 'std', 'min', 'max']].round(2)
```

```python
plotly_pie = px.pie(data_frame=df_raw, names="origin", labels="origin", height=250)
plotly_pie.update_layout(
    showlegend=False,
#     modebar=False,
)
plotly_pie.update_traces(textposition='inside', textinfo='percent+label');
```

```python
pandas_df = df
```

```python
report = es.Page(title="My Report")
report["Data Analysis"] = es.CardSection(cards_equal=True)
report["Data Analysis"][0][0] = es.Card(title="Pie Chart", children=[plotly_pie, pandas_df])
# report["Data Analysis"][0][0] = {"Pie Chart": [plotly_pie, pandas_df]}
report["Data Analysis"][0] = (
    {"Data": pandas_df},
    {"Pie Chart": plotly_pie}
)
report["Data Analysis"][1] = (
    {"Pie Chart": plotly_pie},
    {"Data": pandas_df}
)
report.save_html("my-report.html")
```
